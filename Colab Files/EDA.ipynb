#Modules & Datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Complete_Dataset.csv')
df_state = pd.read_csv('/content/State_Dataset.csv')
df_hotels = pd.read_csv('/content/Hotels.csv')
```

#State Wise Tourism Distribution

```python
df_state.plot(x="States/UTs", y=["2016","2017","2018 (Revised)"], kind="bar",figsize=(20,10))
```

#Circle Wise Tourism Distribution and EDA

```python
df_sum = df.groupby('Circle')[['Domestic - 2016-17', 'Foreign - 2016-17','Domestic - 2017-18','Foreign - 2017-18']].sum().reset_index()
df_sum.head()
```

```python
df_sum.shape
```

```python
df_sum.isnull().sum()
```

```python
medianDomestic_Percentage = df['Domestic Percentage growth - 2017-18/2016-17'].mean()
medianForeign_Percentage = df['Foreign Percentage growth - 2017-18/2016-17'].mean()

df['Domestic Percentage growth - 2017-18/2016-17'].replace(np.nan,medianDomestic_Percentage, inplace = True )
df['Foreign Percentage growth - 2017-18/2016-17'].replace(np.nan,medianForeign_Percentage, inplace = True )
```

```python
df.boxplot(column = ['Airport','Railway','Domestic - 2016-17','Foreign - 2016-17','Domestic - 2017-18','Foreign - 2017-18','Domestic Percentage growth - 2017-18/2016-17','Foreign Percentage growth - 2017-18/2016-17'],figsize=(20,10))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(df.corr(), center=0, annot=True )
ax.set_title('Visulaization')
```

```python
df_sum.plot(x="Circle", y=['Domestic - 2016-17','Foreign - 2016-17','Domestic - 2017-18','Foreign - 2017-18'], kind="bar",figsize=(20,10))
```

```python
plt.figure(figsize=(15,10))
sns.countplot(y='Circle',data=df)
plt.title('No of Tourism Places Counts (City Wise)')
plt.show()
```

```python
plt.figure(1, figsize=(30,5))
n = 0
for x in ['Domestic - 2016-17','Foreign - 2016-17','Domestic - 2017-18','Foreign - 2017-18']:
  n += 1
  plt.subplot(1,4,n)
  plt.subplots_adjust(hspace = 2.5 , wspace = 0.5)
  sns.boxplot(df_sum[x])

plt.show()
```

```python
plt.figure(1, figsize=(30,5))
n = 0
for x in ['Domestic - 2016-17','Foreign - 2016-17','Domestic - 2017-18','Foreign - 2017-18']:
  n += 1
  plt.subplot(1,4,n)
  plt.subplots_adjust(hspace = 2.5 , wspace = 0.5)
  sns.distplot(df_sum[x],bins = 20)

plt.show()
```
